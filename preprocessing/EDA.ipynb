```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv('all_data.csv')
df.head()
```

```python
df.describe().T
```

```python
# Print the percentage of missing values for each column
df.isnull().sum() / len(df) * 100
```

```python
# print columns with NaN values for certain column
dummy_df = df[df['Accelerometer_time_(s)'].isnull()]
dummy_df
```

```python
# Plotting the time series for the accelerometer data
df.plot(x='Accelerometer_time_(s)', y=['Accelerometer_acceleration_x_(m/s^2)',
                                      'Accelerometer_acceleration_y_(m/s^2)',
                                      'Accelerometer_acceleration_z_(m/s^2)'])

# Adding a legend
plt.legend()

# Displaying the plot
plt.show()
```
